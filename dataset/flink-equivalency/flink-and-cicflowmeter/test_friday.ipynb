```python
distinguishing_keys = [
    'Src IP',
    'Src Port',
    'Dst IP',
    'Dst Port',
    'Timestamp',
    'Total Fwd Packet',
    'Total Bwd packets'
]
```

```python
flink_keys = flink[distinguishing_keys + ['Protocol']]
has_duplicate = (flink_keys.duplicated(keep=False)).any()
print(f"Has Duplicates? {has_duplicate}")

flink_unique = flink[~flink_keys.duplicated(keep=False)]
flink_unique
```

```python
real_keys = real[distinguishing_keys + ['Protocol']]
has_duplicate = (real_keys.duplicated(keep=False)).any()
print(f"Has Duplicates? {has_duplicate}")

real_unique = real[~real_keys.duplicated(keep=False)]
real_unique
```

```python
test = real_keys.reset_index().merge(flink_keys.reset_index(), on=distinguishing_keys + ['Protocol'], how='outer', indicator=True)
```
