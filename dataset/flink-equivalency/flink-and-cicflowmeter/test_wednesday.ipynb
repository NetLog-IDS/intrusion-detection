```python
distinguishing_keys = [
    'Src IP',
    'Src Port',
    'Dst IP',
    'Dst Port',
    'Timestamp',
]
```

```python
flink_keys = flink[distinguishing_keys]
has_duplicate = (flink_keys.duplicated(keep=False)).any()
print(f"Has Duplicates? {has_duplicate}")

flink_unique = flink[~flink_keys.duplicated(keep=False)]
flink_unique
```

```python
real_tcp_keys = real_tcp[distinguishing_keys]
has_duplicate = (real_tcp_keys.duplicated(keep=False)).any()
print(f"Has Duplicates? {has_duplicate}")

real_tcp_unique = real_tcp[~real_tcp_keys.duplicated(keep=False)]
real_tcp_unique
```

```python
test = real_tcp_keys.reset_index().merge(flink_keys.reset_index(), on=distinguishing_keys, how='outer', indicator=True)
```

# Problem 1: Wrong Flow ID (Flipped)

```python
pairs = test[test['_merge'] == 'both'][['index_x', 'index_y']]
pairs['index_x'] = pairs['index_x'].astype(int)
pairs['index_y'] = pairs['index_y'].astype(int)
pairs
```

```python
pair_left = list(pairs['index_x'])
pair_right = list(pairs['index_y'])
```

```python
real_tcp_sorted = real_tcp.loc[pair_left].reset_index(drop=True)
real_tcp_sorted
```

```python
flink_sorted = flink.loc[pair_right].reset_index(drop=True)
flink_sorted
```
